```python
df.columns   #it is used to see all the columns in the dataset
```

```python
columns_keep=['Age', 'Gender', 'Marital Status', 'Occupation', 'Monthly Income',
       'Educational Qualifications', 'Family size','Output', 'Feedback']
columns_drop=['latitude', 'longitude',
       'Pin code','Unnamed: 12']
```

```python
df=df[columns_keep]
```

```python
df
```

```python
df.columns
```

```python
df.rename(columns={"Monthly Income":"Monthly salary","Educational Qualifications":"Qualification"},inplace=True) #renaming the columns
```

```python
df
```

```python
df.duplicated().sum() #checking the duplicated values
```

```python
df.drop_duplicates(inplace=True) #droping the duplicate values
```

```python
df
```

```python
df.duplicated().sum()
```

```python
  df.isna().sum() #checking the null values
```

```python
df["Age{months}"]=df["Age"]*12
```

```python
df["Gender"]=df["Gender"].apply(lambda x: "F" if x=="Female" else "M")  #used to change F for Female and M for Male
```

```python
df
```

```python
df["Family size"]=df["Family size"].apply(lambda x:"Large" if x>3 else "Small") #used to change Large is more than 3 orelse small
```

```python
df
```

```python
df["Output"]=df["Output"].apply (lambda x:1 if x=="Yes" else 0) #used to change 1 if its Yes or else 0
```

```python
df["Output"]=df["Output"].astype(float) #changing datatype from int to float
```

```python
df
```

```python
df.reset_index() #resting the index
```
