```python
drop_columns=["unnamed: 0",'LunchType',"EthnicGroup"]
keep_columns=['Gender',"ParentEduc","TestPrep","MathScore",'ReadingScore','WritingScore']
```

```python
df=df[keep_columns]
```

```python
df
```

```python
df.rename(columns={"ParentEduc":"Parent_Education","TestPrep":"Test_Preparation"},inplace=True) #renaming the column names
```

```python
df
```

```python
df.duplicated().value_counts() #remove the duplicates
```

```python
df.drop_duplicates(inplace=True) #droping the duplicates 
```

```python
df.duplicated().sum() 
```

```python
df=df.reset_index().drop(columns="index") # reseting the index
```

```python
df
```

```python
df.isna().sum() #checking wheater their is null values or not
```

```python
df["Gender"]=df["Gender"].apply(lambda x: "F" if x=="female" else "M") # using lambda to change from female and male to "f" and "m" 
```

```python
df["Total_Score"]= df["ReadingScore"]+df["WritingScore"]+df["MathScore"]# total score added of reading score,writing score,mathscore
```

```python
df
```

```python
df["Success"]=df["Total_Score"].apply (lambda x: "Pass" if x >210 else "Fail") #total score is above 210 pass or fail
```
